# Analysis of simulation results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Data cleaning

```python
batch_df = pd.read_csv("results/results_diff_params.csv")

# Split batch_df into the 4 simulations
grouped = batch_df.groupby("RunId")
result = {}
for RunId, group in grouped:
    result[RunId] = group


df1 = result[0]
df2 = result[1]
df3 = result[2]
df4 = result[3]
```

```python
# Figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(20, 5))

legend_labels = ['Overall Users Unable to Find Seat', 'Current Users on Floor']

for i, df in enumerate([df1, df2, df3, df4]):
    df['time'] = pd.to_datetime(df['time'])
    
    # Density curve for 'overall_num_users_unable_to_find_seat' 
    axs[i // 2, i % 2].plot(df['time'], df['overall_num_users_unable_to_find_seat'], label=legend_labels[0], linestyle='-', marker='o', markersize=4)
    axs[i // 2, i % 2].fill_between(df['time'], 0, df['overall_num_users_unable_to_find_seat'], alpha=0.5)
    
    # Density curve for 'current_num_users_on_floor' 
    axs[i // 2, i % 2].plot(df['time'], df['current_num_users_on_floor'], label=legend_labels[1], linestyle='-', marker='o', markersize=4)
    axs[i // 2, i % 2].fill_between(df['time'], 0, df['current_num_users_on_floor'], alpha=0.5)

    axs[i // 2, i % 2].set_title(f'Simulation {i+1}')
    axs[i // 2, i % 2].set_xlabel('Time')
    axs[i // 2, i % 2].set_ylabel('Count')
    axs[i // 2, i % 2].grid(True)
    axs[i // 2, i % 2].legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
legend_labels = ['Simulation 1', 'Simulation 2', 'Simulation 3', 'Simulation 4']

# Iterate through dataframes and plot each one
for i, df in enumerate([df1, df2, df3, df4]):
    df['time'] = pd.to_datetime(df['time'])
    
    # Plot a line for 'current_num_users_on_floor'
    ax.plot(df['time'], df['current_num_users_on_floor'], label=legend_labels[i], linestyle='-', marker='o', markersize=4)

ax.set_title('Current Users on Floor Over Time')
ax.set_xlabel('Time')
ax.set_ylabel('Count')
ax.grid(True)
ax.legend()

# Rotate x-axis labels
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
