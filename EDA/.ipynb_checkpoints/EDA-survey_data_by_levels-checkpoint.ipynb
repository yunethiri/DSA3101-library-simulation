```python
level_df = survey_data[['Level 3 Preference', 'Level 4 Preference', 'Level 5 Preference', 'Level 6 CLB Preference', 'Level 6 Chinese Lib Preference']]
print(level_df)
```

```python
# cleaning data format, standardise numerical format
replace_dict ={"1 - Least preferred" : 1, 
"2" : 2, 
"3" : 3, 
"4" : 4, 
"5 - Most preferred" : 5, 
"NaN" : 0}

for i in level_df.columns:
    level_df = level_df.map(lambda x: replace_dict.get(x, x))

level_df = level_df.fillna(0) # replace NaN with 0

def count_max_occurrences(df):
    column_counts = {col: 0 for col in df.columns}

    for _, row in df.iterrows():
        max_value = row.max()
        max_columns = row[row == max_value].index
        for col in max_columns:
            column_counts[col] += 1

    return column_counts

column_counts = count_max_occurrences(level_df)

print(column_counts)
print(level_df)
```

```python
# Plot by average scores per level
averages = level_df.mean()

# Create a bar plot
ax = averages.plot(kind='bar', figsize=(10, 6))

# Annotate the bars with their respective values
for i in ax.patches:
    ax.annotate(format(i.get_height(), '.2f'), (i.get_x() + i.get_width() / 2, i.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Set labels and title
plt.xlabel('Levels')
plt.ylabel('Averag Preference Score')
plt.title('Average Preference Scores per Level')

# Show the plot
plt.show()
```

The distribution of preferences across all levels are relatively even. Out of all the levels, Level 4 is the most popular and Level 6 Chinese Library is the least popular.
