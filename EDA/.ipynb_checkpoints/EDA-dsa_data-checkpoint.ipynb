<font size="5"> Average Time Spent In Library Per Student </font>

```python
def clear_mismatched_entries_exits(df): 
# only works for non-exam period because doesnt take entry/exit into account
    
    missing_exits_df = pd.DataFrame(columns=df.columns)

    prev_row = None
    for index, row in df.iterrows():
        if prev_row is not None and row['Direction'] == 'Entry':
            if prev_row['Direction'] == 'Entry':
            # Entry followed by Entry, add prev row to the missing_exits_df
                missing_exits_df = pd.concat([missing_exits_df, prev_row.to_frame().T], ignore_index=True)
        elif prev_row is not None and prev_row['Direction'] == 'Exit' and row['Direction'] == 'Exit':
            # Entry followed by Exit, add row to missing_exits_df
            missing_exits_df = pd.concat([missing_exits_df, row.to_frame().T], ignore_index=True)
        elif prev_row is not None and prev_row['Direction'] == 'Entry' and row['Direction'] == 'Exit':
            # Entry followed by Exit or Entry/Exit, no action needed
            if prev_row['User Number'] != row['User Number']:
                missing_exits_df = pd.concat([missing_exits_df, prev_row.to_frame().T], ignore_index=True)
                missing_exits_df = pd.concat([missing_exits_df, row.to_frame().T], ignore_index=True)
            elif prev_row['User Number'] == row['User Number']:
                pass
        prev_row = row
        
    # Check if the last row is a mismatched entry or exit
    if prev_row is not None and row['Direction'] == 'Entry':
        missing_exits_df = pd.concat([missing_exits_df, row.to_frame().T], ignore_index=True)

    # New dfs with rows in missing_exits_df removed
    new_df = df[~df[['User Number', 'Datetime']].apply(tuple, axis=1).isin(missing_exits_df[['User Number', 'Datetime']].apply(tuple, axis=1))].reset_index(drop = True)
    
    return new_df
```

```python
def avg_time_spent_per_student_non_exam(df):
# only works for non-exam period because doesnt take entry/exit into account

    df = clear_mismatched_entries_exits(df)
    
    # Separate DataFrames based on 'Direction'
    entry_df = df[df['Direction'] == 'Entry'].reset_index()
    exit_df = df[df['Direction'] == 'Exit'].reset_index()
    exit_df.columns = [f'{col}_exit' for col in exit_df.columns]
    entry_exit = pd.concat((entry_df, exit_df), axis=1)

    entry_exit['Time'] = pd.to_datetime(entry_exit['Time'], format='%H:%M:%S')
    entry_exit['Time_exit'] = pd.to_datetime(entry_exit['Time_exit'], format='%H:%M:%S')

    # Calculate time spent
    entry_exit['Time Spent'] = entry_exit['Time_exit'] - entry_exit['Time']

    # Group by 'User Number' and calculate the mean time spent
    df = entry_exit.groupby('User Number')['Time Spent'].sum().reset_index() 
    # to discuss either mean or sum () -> mean() can find turnover rate in general while
    # sum can find how long each student spends in the library excluding when they go for 
    # lunch or classes etc. 
    df['Time Spent'] = pd.to_timedelta(df['Time Spent'])
    df['Hour'] = df['Time Spent'].dt.total_seconds() / 3600  # Convert to hours
    df['Hour'] = df['Hour'].round() 
    average_time_spent_per_student = df
    # Calculate the average time spent per day
    #average_time_spent_day = df.groupby('User Number').size().mean()

    return average_time_spent_per_student
```

```python
# for reading week and exam week 
def remove_user_numbers(df): # remove user numbers which only has one row
    grouped = df.groupby('User Number')
    filtered_groups = [group for _, group in grouped if len(group) > 1]
    if filtered_groups:
        new_df = pd.concat(filtered_groups)
    else:
        new_df = pd.DataFrame(columns=df.columns)
        
    return new_df

def avg_time_spent_per_student_exam(df):
    
    df = remove_user_numbers(df)

    unique_user_numbers = df['User Number'].unique()

    overall_duration_df = pd.DataFrame()

    for user_number in unique_user_numbers:
        user_df = df[df['User Number'] == user_number].copy()
        prev_row = None
        entry_time_store = None
        rows_left = len(user_df)
        current_row = 1
        durations_df = pd.DataFrame()
        user_df['Time'] = pd.to_datetime(user_df['Time'], format='%H:%M:%S')

        for index, row in user_df.iterrows():
            if prev_row is not None and prev_row['Direction'] == 'Entry':
                if row['Direction'] == 'Exit':
                    durations_df = pd.concat([durations_df, pd.DataFrame({'User Number': user_number, 'Duration': [(row['Time'] - prev_row['Time'])]})], ignore_index=True)
                elif (rows_left > 2 and current_row%2 == 0) or rows_left == 1:
                    if row['Direction'] == 'Entry/Exit':
                        durations_df = pd.concat([durations_df, pd.DataFrame({'User Number': user_number, 'Duration': [(row['Time'] - prev_row['Time'])]})], ignore_index=True)
                elif rows_left == 2:
                    if row['Direction'] == 'Entry/Exit':
                        entry_time_store = prev_row['Time']
            elif prev_row is not None and prev_row['Direction'] == 'Entry/Exit':
                if row["Direction"] == "Entry":
                    prev_row = row
                    rows_left -= 1
                    current_row += 1
                    continue
                elif row["Direction"] == "Entry/Exit":
                    if rows_left == 2:
                        entry_time_store = prev_row['Time']
                    elif rows_left == 1 and entry_time_store:
                        durations_df = pd.concat([durations_df, pd.DataFrame({'User Number': user_number, 'Duration': [(row['Time'] - entry_time_store)]})], ignore_index=True)
                    elif (rows_left > 2 and current_row%2 == 0) or rows_left == 1:
                        durations_df = pd.concat([durations_df, pd.DataFrame({'User Number': user_number, 'Duration': [(row['Time'] - prev_row['Time'])]})], ignore_index=True)
                elif row["Direction"] == "Exit":
                    durations_df = pd.concat([durations_df, pd.DataFrame({'User Number': user_number, 'Duration': [(row['Time'] - prev_row['Time'])]})], ignore_index=True)
            prev_row = row
            rows_left -= 1
            current_row += 1

        # Append the durations_df to the overall_duration_df
        overall_duration_df = pd.concat([overall_duration_df, durations_df], ignore_index=True)
        
    # Group by 'User Number' and sum the 'Duration' for each user
    df = overall_duration_df.groupby('User Number')['Duration'].sum().reset_index()

    # Extract hours from the 'Duration'
    df['Hour'] = df['Duration'].dt.total_seconds() / 3600  # Convert to hours
    df['Hour'] = df['Hour'].round() 

    overall_duration_df = df
        
    # Now, overall_duration_df contains user numbers and their calculated durations
    return overall_duration_df
```

```python
# Define your data
period1 = ['2023-01-25', '2023-01-26', '2023-01-27']  # week 3
period2 = ['2023-02-18', '2023-02-20', '2023-02-21', '2023-02-22', '2023-02-23', '2023-02-24', '2023-02-25']  # recess week
period3 = ['2023-04-15', '2023-04-17', '2023-04-18', '2023-04-19', '2023-04-20', '2023-04-21']  # reading week
period4 = ['2023-04-22', '2023-04-24', '2023-04-25', '2023-04-26', '2023-04-27', '2023-04-28', '2023-04-29']  # exam week

# Create a list of periods
periods = [period1, period2, period3, period4]

# Define other parameters
students = ['Undergraduate', 'Postgraduate', 'Special Student', 'SCALE Students']
df = dsa_data

# Create a list to store DataFrames for each period
period_data_frames = []

# Iterate through periods
for i, day_list in enumerate(periods):
    if i == 0 or i == 1:
        function = avg_time_spent_per_student_non_exam
    elif i == 2 or i == 3:
        function = avg_time_spent_per_student_exam

    # Create a DataFrame to store the data for the current period
    period_df = pd.DataFrame(columns=['Hour', 'Number of Students'])

    # Iterate through days within the current period
    for day in day_list:
        # Filter the DataFrame for the current day within the period
        day_df = df[(df['Date'] == day) & (df['Broad Category'].isin(students))]
        day_df = day_df.sort_values(by=["User Number", "Time"])

        day_df = function(day_df)

        # Group data by 'Hour' and calculate the number of students for each hour
        hourly_counts = day_df.groupby('Hour')['User Number'].nunique()

        # Create a DataFrame for the current day's data
        day_data = pd.DataFrame({'Hour': hourly_counts.index, 'Number of Students': hourly_counts.values})

        # Append the current day's data to the period DataFrame
        period_df = pd.concat([period_df, day_data], ignore_index=True)

    # Append the period DataFrame to the list
    period_data_frames.append(period_df)

for i, period_df in enumerate(period_data_frames):
    period_df['Number of Students'] = period_df['Number of Students'].astype(int)  # Convert the column to integer
    summed_data = period_df.groupby('Hour')['Number of Students'].sum().reset_index()
    # Sum up the number of students for each hour
    overall_total = summed_data['Number of Students'].sum()

    # Calculate the percentages for each hour
    summed_data['Percentage'] = (summed_data['Number of Students'] / overall_total) * 100

    # Print the DataFrame with percentages
    #print(summed_data)


hour_percentage_dicts = []

for i, period_df in enumerate(period_data_frames):
    print(f"Period {i + 1} DataFrame:")
    period_df['Number of Students'] = period_df['Number of Students'].astype(int)  # Convert the column to an integer
    summed_data = period_df.groupby('Hour')['Number of Students'].sum().reset_index()

    # Sum up the number of students for each hour
    overall_total = summed_data['Number of Students'].sum()

    # Calculate the percentages for each hour
    summed_data['Percentage'] = (summed_data['Number of Students'] / overall_total) * 100

    hour_percentage_dict = {}

    for index, row in summed_data.iterrows():
        hour = int(row['Hour'])  # Convert the hour to an integer
        percentage = round(row['Percentage'], 3)  # Round the percentage to one decimal place
        hour_percentage_dict[hour] = percentage

    hour_percentage_dicts.append(hour_percentage_dict)

# Print the list of dictionaries
for i, hour_percentage_dict in enumerate(hour_percentage_dicts):
    print(f"Period {i + 1} Dictionary:")
    print(hour_percentage_dict)


# If you want to save each period's DataFrame to a CSV file, you can use result_df.to_csv() here.
```
