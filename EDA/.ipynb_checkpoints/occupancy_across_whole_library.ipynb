```python
import pandas as pd
from datetime import datetime, timedelta, timezone
import matplotlib.pyplot as plt
```

```python
# Read in data
dsa_data = pd.read_csv("../data/dsa_data.csv")
```

# Data Cleaning (Defining timeframes, filtering for only students)

```python
# Data Cleaning 

dsa_data['Datetime'] = pd.to_datetime(dsa_data['Datetime'])

utc_offset = timezone(timedelta(hours=8))

dsa_data['Datetime'] = dsa_data['Datetime'].dt.tz_convert(utc_offset)

# Split by Date, Hour, Time 
dsa_data['Date'] = dsa_data['Datetime'].dt.strftime("%Y-%m-%d")
dsa_data['Hour'] = dsa_data['Datetime'].dt.hour
dsa_data['Time'] = dsa_data['Datetime'].dt.strftime("%H:%M:%S")

#filter for only students 
students = ['Undergraduate', 'Postgraduate','Special Students','SCALE Students']
# Filter the DataFrame based on the 'Broad Category' column
dsa_data = dsa_data[dsa_data['Broad Category'].isin(students)]

# week 3 - only three days (Wed to Fri) as Mon and Tues were CNY public hols
week3_df = dsa_data[(dsa_data['Date'] >= '2023-01-25') & (dsa_data['Date'] <= '2023-01-27')]

# recess week - weekdays only  
recess_week = dsa_data[(dsa_data['Date'] >= '2023-02-20') & (dsa_data['Date'] <= '2023-02-24')]

# reading week - weekdays only
reading_week = dsa_data[(dsa_data['Date'] >= '2023-04-17') & (dsa_data['Date'] <= '2023-04-21')]

# exam week 1 - weekdays only
examweek_one = dsa_data[(dsa_data['Date'] >= '2023-04-24') & (dsa_data['Date'] <=  '2023-04-28')]
```

# Finding average occupancy across time (hour) for each time period


```python
# Create a new 'Count' column based on the 'Direction' column
week3_df['Count'] = 0
week3_df.loc[week3_df['Direction'] == 'Entry', 'Count'] = 1
week3_df.loc[week3_df['Direction'] == 'Exit', 'Count'] = -1

# Group and aggregate the data by the 'Hour' column
hourly_counts = week3_df.groupby('Hour')['Count'].sum().cumsum()

# Divide the cumulative counts by 3 (Average over the 3 days)
hourly_counts /= 3

# Plot the adjusted cumulative counts
plt.figure(figsize=(12, 6))
hourly_counts.plot(kind='line', title='Average Occupancy in the Library Over Hours (No Exams Period)')
plt.xlabel('Hour')
plt.ylabel('Adjusted Cumulative People Count')
plt.grid()
# Add annotations with hour labels
for hour, count in hourly_counts.items():
    plt.text(hour, count, str(int(count)), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Create a new 'Count' column based on the 'Direction' column for reading_week_data
recess_week['Count'] = 0
recess_week.loc[recess_week['Direction'] == 'Entry', 'Count'] = 1
recess_week.loc[recess_week['Direction'] == 'Exit', 'Count'] = -1

start_hour = 7


# Group and aggregate the data by the 'Hour' column
hourly_counts_rw = recess_week.groupby('Hour')['Count'].sum().cumsum()

# Divide the cumulative counts by 5 (Average over the 5 days)
hourly_counts_rw /= 5
hourly_counts_rw = hourly_counts_rw[hourly_counts_rw.index >= start_hour]

# Plot the adjusted cumulative counts for recess_week
plt.figure(figsize=(12, 6))
hourly_counts_rw.plot(kind='line', title='Average Occupancy in the Library Over Hours (Recess Week)')
plt.xlabel('Hour')
plt.ylabel('Adjusted Cumulative People Count')
plt.grid()

# Add annotations with hour labels for recess_week
for hour, count in hourly_counts_rw.items():
    plt.text(hour, count, str(int(count)), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Create a new 'Count' column based on the 'Direction' column for reading_week
reading_week['Count'] = 0
reading_week.loc[reading_week['Direction'] == 'Entry', 'Count'] = 1
reading_week.loc[reading_week['Direction'] == 'Exit', 'Count'] = -1

start_hour = 7

# Group and aggregate the data by the 'Hour' column
hourly_counts_reading_week = reading_week.groupby('Hour')['Count'].sum().cumsum()

# Divide the cumulative counts by 5 (Average over the 5 days)
hourly_counts_reading_week /= 5
hourly_counts_reading_week = hourly_counts_reading_week[hourly_counts_reading_week.index >= start_hour]

# Plot the adjusted cumulative counts for reading_week
plt.figure(figsize=(12, 6))
hourly_counts_reading_week.plot(kind='line', title='Average Occupancy in the Library Over Hours (Reading Week)')
plt.xlabel('Hour')
plt.ylabel('Adjusted Cumulative People Count')
plt.grid()

# Add annotations with hour labels for reading_week
for hour, count in hourly_counts_reading_week.items():
    plt.text(hour, count, str(int(count)), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Create a new 'Count' column based on the 'Direction' column for examweek_one
examweek_one['Count'] = 0
examweek_one.loc[examweek_one['Direction'] == 'Entry', 'Count'] = 1
examweek_one.loc[examweek_one['Direction'] == 'Exit', 'Count'] = -1

start_hour = 7

# Group and aggregate the data by the 'Hour' column
hourly_counts_exam_week_one = examweek_one.groupby('Hour')['Count'].sum().cumsum()

# Divide the cumulative counts by 5 (Average over the 5 days)
hourly_counts_exam_week_one /= 5
hourly_counts_exam_week_one = hourly_counts_exam_week_one[hourly_counts_exam_week_one.index >= start_hour]

# Plot the adjusted cumulative counts for examweek_one
plt.figure(figsize=(12, 6))
hourly_counts_exam_week_one.plot(kind='line', title='Average Occupancy in the Library Over Hours (Exam Week One)')
plt.xlabel('Hour')
plt.ylabel('Adjusted Cumulative People Count')
plt.grid()

# Add annotations with hour labels for examweek_one
for hour, count in hourly_counts_exam_week_one.items():
    plt.text(hour, count, str(int(count)), ha='center', va='bottom', fontsize=10)

plt.show()
```

# Finding unique number of users per day

```python
# Group and aggregate the data in dsa_data to count unique User Numbers per Date
unique_users_per_date_dsa = dsa_data.groupby('Date')['User Number'].nunique()

# Filter rows with NA User Numbers and divide by 2 (Assume those who scan using QR code enter and exit once)
na_rows_dsa = dsa_data['User Number'].isna().groupby(dsa_data['Date']).sum() / 2

# Add the NA count to the unique user count for those Dates with NA in dsa_data
unique_users_per_date_dsa += na_rows_dsa

# Print the result for dsa_data
print(unique_users_per_date_dsa)
```
