```python
import pandas as pd
from datetime import datetime, timedelta, timezone
import matplotlib.pyplot as plt
```

```python
# Read in data

# first dataset 
taps_data = pd.read_csv("../data/20230413_clb_taps.csv")
gates_data = pd.read_excel("../data/cl_gates.xlsx")
```

First Dataset 

```python
taps_data.head()

# date of the data is 13 April 2023 -> Exam Period
```

```python
# Data Cleaning 

taps_data['Datetime'] = pd.to_datetime(taps_data['Datetime'])

utc_offset = timezone(timedelta(hours=8))

taps_data['Datetime'] = taps_data['Datetime'].dt.tz_convert(utc_offset)

# Split by Date, Hour, Time 
taps_data['Date'] = taps_data['Datetime'].dt.strftime("%Y-%m-%d")
taps_data['Hour'] = taps_data['Datetime'].dt.hour
taps_data['Time'] = taps_data['Datetime'].dt.strftime("%H:%M:%S")
```

```python
overall_df = pd.merge(taps_data, gates_data, on='Device Code', how = 'left')
```

```python
overall_df.head()
```

```python
# Group by 'Device code' and count the entries and exits separately
device_code_counts = overall_df.groupby(['Device Code', 'Direction']).size().unstack(fill_value=0).reset_index()

# Calculate total entries for each 'Device code' and sort 
device_code_counts['Total Entries'] = device_code_counts['Entry'] + device_code_counts['Exit']

# Plot the data
plt.figure(figsize=(12, 6))
# Entry 
plt.bar(device_code_counts['Device Code'], device_code_counts['Entry'], color='blue', label='Entry')
# Exit
plt.bar(device_code_counts['Device Code'], device_code_counts['Exit'], color='red', label='Exit', bottom=device_code_counts['Entry'])

plt.xlabel('Device Code')
plt.ylabel('Number of Gantry Taps')
plt.title('Number of Entries and Exits by Device Code')
plt.xticks(rotation=45, ha='right') 
plt.legend()
plt.tight_layout()
plt.show()

'''
Observations:
1) Entry and Exit is highest via QR Code on Level 1 Gantry 
During this time period (date of the data is 13 April 2023 -> Exam Period),
2) top 3 highest entry taps is via Level 1 or Level 6 
3) Highest exit taps are via Level 1 or Level 6
-> gets more crowded at level 6 gantries during exam period due to level 5 and level 6 study tables

'''

```

```python
access_type = overall_df.groupby('Access Type').size().reset_index(name='Number of Taps')

# Plot the data
plt.figure(figsize=(12, 6))

plt.bar(access_type['Access Type'], access_type['Number of Taps'], color='blue')

plt.xlabel('Access Type')
plt.ylabel('Number of Gantry Taps')
plt.title('Number of Gantry Taps by Access Type')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
taps_data['Hour'] = taps_data['Datetime'].dt.hour

hourly_counts = taps_data.groupby(['Hour', 'Direction']).size().unstack(fill_value=0).reset_index()
hourly_counts['Total Entries'] = hourly_counts['Entry'] + hourly_counts['Exit']
hourly_counts = hourly_counts.sort_values(by='Hour')

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(hourly_counts['Hour'], hourly_counts['Entry'], color='blue', label='Entry', alpha=0.7)
plt.bar(hourly_counts['Hour'], hourly_counts['Exit'], color='red', label='Exit', alpha=0.7)

plt.xlabel('Hour of the Day')
plt.ylabel('Number of Entries and Exits')
plt.title('Number of Entries and Exits Hour by Hour')
plt.legend()
plt.tight_layout()
plt.show()

'''
Observations: 
1) More Entries from 9 am to 2 pm 
2) More Exits from 3 pm onwards with the largest peak at 5 pm for dinner time 
3) Consistently a large number of exits between 11 am to 2 pm -> lunch time 

'''
```
